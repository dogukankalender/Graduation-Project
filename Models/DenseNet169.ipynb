```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers, Model, optimizers
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet169
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns

train_yes_path = '/content/datadenek2/beyin/veri/yes1'
train_no_path = '/content/datadenek2/beyin/veri/no1'

data = []
labels = []

for r, d, f in os.walk(train_yes_path):
    for file in f:
        if file.endswith('.jpg'):
            img_path = os.path.join(r, file)
            img = Image.open(img_path)
            img = img.resize((128, 128))
            img = np.array(img)
            if img.shape == (128, 128, 3):
                data.append(img)
                labels.append(0)

for r, d, f in os.walk(train_no_path):
    for file in f:
        if file.endswith('.jpg'):
            img_path = os.path.join(r, file)
            img = Image.open(img_path)
            img = img.resize((128, 128))
            img = np.array(img)
            if img.shape == (128, 128, 3):
                data.append(img)
                labels.append(1)

if len(data) == 0 or len(labels) == 0:
    raise ValueError("Veriler veya etiketler boÅŸ. LÃ¼tfen veri yÃ¼kleme sÃ¼recini kontrol edin.")

data = np.array(data)
labels = np.array(labels)

encoder = OneHotEncoder()
labels = encoder.fit_transform(labels.reshape(-1, 1)).toarray()

x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, shuffle=True, random_state=0)

x_train = x_train / 255.0
x_test = x_test / 255.0

train_datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest',
)

train_datagen.fit(x_train)

base_model = DenseNet169(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

for layer in base_model.layers:
    layer.trainable = False

for layer in base_model.layers[-20:]:
    layer.trainable = True

x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(2, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=outputs)

model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

checkpoint = ModelCheckpoint("best_model.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1)

early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)

history = model.fit(
    train_datagen.flow(x_train, y_train, batch_size=32),
    steps_per_epoch=len(x_train) / 32,
    epochs=50,
    validation_data=(x_test, y_test),
    callbacks=[checkpoint, lr_scheduler, early_stopping]
)

evaluation_results = model.evaluate(x_test, y_test)
accuracy = evaluation_results[1]
print("Test Setinde Model DoÄŸruluÄŸu: {:.2f}%".format(accuracy * 100))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model KayÄ±p')
plt.ylabel('KayÄ±p')
plt.xlabel('Epoch')
plt.legend(['EÄŸitim', 'DoÄŸrulama'], loc='upper right')
plt.show()

def names(number):
    if number == 0:
        return 'bir tÃ¼mÃ¶rdÃ¼r'
    else:
        return 'bir tÃ¼mÃ¶r deÄŸildir'

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

precision = precision_score(y_true_classes, y_pred_classes)
recall = recall_score(y_true_classes, y_pred_classes)
f1 = f1_score(y_true_classes, y_pred_classes)

print("Kesinlik: {:.2f}".format(precision))
print("HatÄ±rlama: {:.2f}".format(recall))
print("F1 Skoru: {:.2f}".format(f1))

cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['tÃ¼mÃ¶r bulunan', 'tÃ¼mÃ¶r bulunmayan'], yticklabels=['tÃ¼mÃ¶r bulunan', 'tÃ¼mÃ¶r bulunmayan'])
plt.ylabel('GerÃ§ek Etiket')
plt.xlabel('Tahmin Edilen Etiket')
plt.title('KarÄ±ÅŸÄ±klÄ±k Matrisi')
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers, Model, optimizers
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet169
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns

# EÄŸitim verileri iÃ§in yollarÄ± ayarlayÄ±n
train_yes_path = '/content/datadenek2/beyin/veri/yes1'
train_no_path = '/content/datadenek2/beyin/veri/no1'

# GÃ¶rselleri yÃ¼kleyin ve iÅŸleyin
data = []
labels = []

# 'Yes' (tÃ¼mÃ¶r var) gÃ¶rsellerini iÅŸleyin
for r, d, f in os.walk(train_yes_path):
    for file in f:
        if file.endswith('.jpg'):
            img_path = os.path.join(r, file)
            img = Image.open(img_path)
            img = img.resize((128, 128))
            img = np.array(img)
            if img.shape == (128, 128, 3):
                data.append(img)
                labels.append(0)  # 'yes' sÄ±nÄ±fÄ±

# 'No' (tÃ¼mÃ¶r yok) gÃ¶rsellerini iÅŸleyin
for r, d, f in os.walk(train_no_path):
    for file in f:
        if file.endswith('.jpg'):
            img_path = os.path.join(r, file)
            img = Image.open(img_path)
            img = img.resize((128, 128))
            img = np.array(img)
            if img.shape == (128, 128, 3):
                data.append(img)
                labels.append(1)  # 'no' sÄ±nÄ±fÄ±

# Veri ve etiketlerin doÄŸru doldurulup doldurulmadÄ±ÄŸÄ±nÄ± kontrol edin
if len(data) == 0 or len(labels) == 0:
    raise ValueError("Veriler veya etiketler boÅŸ. LÃ¼tfen veri yÃ¼kleme sÃ¼recini kontrol edin.")

data = np.array(data)
labels = np.array(labels)

# Etiketleri one-hot kodlayÄ±n
encoder = OneHotEncoder()
labels = encoder.fit_transform(labels.reshape(-1, 1)).toarray()

# Veriyi %70 eÄŸitim ve %30 test olarak ayÄ±rÄ±n
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, shuffle=True, random_state=0)

# GÃ¶rÃ¼ntÃ¼ verilerini normalize edin
x_train = x_train / 255.0
x_test = x_test / 255.0

# Veri artÄ±rma
train_datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=False,  # Dikey Ã§evirme genellikle daha az kullanÄ±lÄ±r Ã§Ã¼nkÃ¼ gÃ¶rÃ¼ntÃ¼lerin doÄŸal yapÄ±sÄ±nÄ± bozabilir.
    fill_mode='nearest',
)

train_datagen.fit(x_train)

# DenseNet169 modelini tanÄ±mlayÄ±n
base_model = DenseNet169(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# TÃ¼m katmanlarÄ± dondurun
for layer in base_model.layers:
    layer.trainable = False

# Son 10 katmanÄ± yeniden eÄŸitilebilir yapÄ±n
for layer in base_model.layers[-20:]:
    layer.trainable = True

# DenseNet169 Ã¼zerine Ã¶zel katmanlar ekleyin
x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(2, activation='sigmoid')(x)

# Tam modeli oluÅŸturun
model = Model(inputs=base_model.input, outputs=outputs)

# Modeli derleyin
model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(learning_rate=0.0001), metrics=['accuracy'])

# Model Checkpoint
checkpoint = ModelCheckpoint("best_model.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

# Ã–ÄŸrenme OranÄ± AzaltÄ±cÄ±
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1)

# Erken Durdurma
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)

# Modeli eÄŸitin
history = model.fit(
    train_datagen.flow(x_train, y_train, batch_size=32),
    steps_per_epoch=len(x_train) / 32,
    epochs=50,
    validation_data=(x_test, y_test),
    callbacks=[checkpoint, lr_scheduler, early_stopping]
)

# Modeli deÄŸerlendirin
evaluation_results = model.evaluate(x_test, y_test)
accuracy = evaluation_results[1]
print("Test Setinde Model DoÄŸruluÄŸu: {:.2f}%".format(accuracy * 100))

# KayÄ±p grafiÄŸi Ã§izme
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model KayÄ±p')
plt.ylabel('KayÄ±p')
plt.xlabel('Epoch')
plt.legend(['EÄŸitim', 'DoÄŸrulama'], loc='upper right')
plt.show()

# SÄ±nÄ±f isimleri iÃ§in yardÄ±mcÄ± fonksiyon
def names(number):
    if number == 0:
        return 'bir tÃ¼mÃ¶rdÃ¼r'
    else:
        return 'bir tÃ¼mÃ¶r deÄŸildir'

# Kesinlik, HatÄ±rlama ve F1 Skorunu HesaplayÄ±n
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

precision = precision_score(y_true_classes, y_pred_classes)
recall = recall_score(y_true_classes, y_pred_classes)
f1 = f1_score(y_true_classes, y_pred_classes)

print("Kesinlik: {:.2f}".format(precision))
print("HatÄ±rlama: {:.2f}".format(recall))
print("F1 Skoru: {:.2f}".format(f1))

# KarÄ±ÅŸÄ±klÄ±k Matrisi
cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['tÃ¼mÃ¶r bulunan', 'tÃ¼mÃ¶r bulunmayan'], yticklabels=['tÃ¼mÃ¶r bulunan', 'tÃ¼mÃ¶r bulunmayan'])
plt.ylabel('GerÃ§ek Etiket')
plt.xlabel('Tahmin Edilen Etiket')
plt.title('KarÄ±ÅŸÄ±klÄ±k Matrisi')
plt.show()
```
